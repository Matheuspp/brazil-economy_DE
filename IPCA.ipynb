```python
df = df[df.inflacao < 900]
```

```python
# ploting bars
fig, ax = plt.subplots(figsize=(15, 7))
ax.bar(np.arange(1995, 2020).astype(np.str), np.flipud(df.T.values[1]), color=['gray'], edgecolor=['silver'] , linewidth=1)
ax.set_ylabel("inflation rate")
ax.set_xlabel("Year")
fig.tight_layout()
plt.savefig('graphs/inflation_bar_graph')
```
